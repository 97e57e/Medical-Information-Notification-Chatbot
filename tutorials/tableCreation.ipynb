### 개수 카운팅 

```python
def countings (cate, n):
    cnt = 0
    for a in cate.items():
        if( a[1]>=n):
            cnt +=1
    print ( "%d개 이상인 기관 : %d" %(n,cnt))

countings(cate,1)
countings(cate,2)
countings(cate,3)
countings(cate,4)
```
